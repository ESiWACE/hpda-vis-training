# ESiWACE2 Online Training on High-Performance Data Anlytics (HPDA) and Visualisation

### Tuesday, 6 October, 2020

This online training course aims to increase scientists’ expertise on data analysis and visualization applied to climate and weather domains, using high-performance data analytics and visualization tools available from the open source market (i.e., Ophidia and ParaView). This course is organized in the context of the “Centre of Excellence in Simulation of Weather and Climate in Europe” (ESiWACE) phase 2 project.

## Introduction to Ophidia

This notebook provides the outline for the first session of the training course. The session is organized as:

1. Introduction to HPDA, data challenges in eScience and the Ophidia HPDA framework
2. Practical overview of data analytics with the Ophidia framework (PyOphidia)
3. Hands-on tutorial with some example notebooks with PyOphidia

### Demo Notebook

This folder contains the notebook that will be shown during the practical overview demo:

### Hands-on Notebooks

In this folder you can find the notebooks for the hands-on tutorial:

1. Start from [**Basics.ipynb**](./Hands-on/1-Basics.ipynb) notebook, which provides step-by-step instructions on how to use the Ophidia operators as well as the PyOphidia library to implement a climate indicator. 
2. The second notebook [**Summer_Days.ipynb**](./Hands-on/2-Summer_Days.ipynb), provides a simple exercise for the implementation of a climate indicator with PyOphidia
3. The final notebook [**Linear_Regression.ipynb**](./Hands-on/3-Linear_Regression.ipynb) represents a more challenging exercise, where you can implement a linear regression experiment to extract the trend from the data.

### Example Notebooks

This folder contains a set of ready-to-use (bonus) noteboks that can be executed if you are interested in seeing additional ways of exploiting Ophidia.

- [**Aggregated map**](./Example_notebooks/Aggregated_map.ipynb) shows how to create a map from an Ophidia datacube;
- [**Frost Days**](./Example_notebooks/Frost_Days.ipynb) shows how to compute the Frost Days indicator;
- [**Linear_regression**](./Example_notebooks/Linear_regression.ipynb) provides the user with an exercise regarding linear regression to be performed with the ECAS features and resources;
- [**Summer Days**](./Example_notebooks/Summer_Days.ipynb) shows how to compute the Summer Days indicator;
